```python
import tensorflow as tf
from tensorflow.keras.datasets import mnist, cifar10
from tensorflow.keras.utils import to_categorical
import numpy as np

# Load datasets
(x_train_mnist, y_train_mnist), (x_test_mnist, y_test_mnist) = mnist.load_data()
(x_train_cifar, y_train_cifar), (x_test_cifar, y_test_cifar) = cifar10.load_data()

# Normalize
x_train_mnist = x_train_mnist / 255.0
x_test_mnist = x_test_mnist / 255.0

x_train_cifar = x_train_cifar / 255.0
x_test_cifar = x_test_cifar / 255.0

# Reshape MNIST for ANN
x_train_mnist_flat = x_train_mnist.reshape(-1, 28*28)
x_test_mnist_flat = x_test_mnist.reshape(-1, 28*28)

# One-hot encode labels
y_train_mnist_cat = to_categorical(y_train_mnist, 10)
y_test_mnist_cat = to_categorical(y_test_mnist, 10)
y_train_cifar_cat = to_categorical(y_train_cifar, 10)
y_test_cifar_cat = to_categorical(y_test_cifar, 10)


from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# ANN for MNIST
ann_model = Sequential([
    Dense(128, activation='relu', input_shape=(784,)),
    Dense(64, activation='relu'),
    Dense(10, activation='softmax')
])

ann_model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Train ANN
ann_history = ann_model.fit(x_train_mnist_flat, y_train_mnist_cat, epochs=10, validation_data=(x_test_mnist_flat, y_test_mnist_cat))


from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten

# Reshape MNIST for CNN
x_train_mnist_cnn = x_train_mnist.reshape(-1, 28, 28, 1)
x_test_mnist_cnn = x_test_mnist.reshape(-1, 28, 28, 1)

# CNN for MNIST
cnn_model = Sequential([
    Conv2D(32, (3,3), activation='relu', input_shape=(28,28,1)),
    MaxPooling2D(2,2),
    Flatten(),
    Dense(64, activation='relu'),
    Dense(10, activation='softmax')
])

cnn_model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Train CNN
cnn_history = cnn_model.fit(x_train_mnist_cnn, y_train_mnist_cat, epochs=10, validation_data=(x_test_mnist_cnn, y_test_mnist_cat))



import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, classification_report
import seaborn as sns

# Accuracy and loss plots
def plot_history(history, title):
    plt.figure(figsize=(12, 4))
    plt.subplot(1,2,1)
    plt.plot(history.history['accuracy'], label='Train Acc')
    plt.plot(history.history['val_accuracy'], label='Val Acc')
    plt.title(f'{title} - Accuracy')
    plt.legend()

    plt.subplot(1,2,2)
    plt.plot(history.history['loss'], label='Train Loss')
    plt.plot(history.history['val_loss'], label='Val Loss')
    plt.title(f'{title} - Loss')
    plt.legend()
    plt.show()

plot_history(ann_history, "ANN (MNIST)")
plot_history(cnn_history, "CNN (MNIST)")

# Confusion Matrix for CNN
y_pred_cnn = np.argmax(cnn_model.predict(x_test_mnist_cnn), axis=1)
cm = confusion_matrix(y_test_mnist, y_pred_cnn)

plt.figure(figsize=(8,6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.title("Confusion Matrix - CNN (MNIST)")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()

# Classification Report
print("Classification Report - CNN (MNIST):")
print(classification_report(y_test_mnist, y_pred_cnn))




```

```python
# Flatten CIFAR images for ANN
x_train_cifar_flat = x_train_cifar.reshape(-1, 32*32*3)
x_test_cifar_flat = x_test_cifar.reshape(-1, 32*32*3)

# ANN model for CIFAR-10
ann_cifar = Sequential([
    Dense(512, activation='relu', input_shape=(32*32*3,)),
    Dense(256, activation='relu'),
    Dense(10, activation='softmax')
])

ann_cifar.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Train ANN on CIFAR-10
ann_cifar_history = ann_cifar.fit(x_train_cifar_flat, y_train_cifar_cat, epochs=10, validation_data=(x_test_cifar_flat, y_test_cifar_cat))

# CNN model for CIFAR-10
cnn_cifar = Sequential([
    Conv2D(32, (3,3), activation='relu', input_shape=(32,32,3)),
    MaxPooling2D(2,2),
    Conv2D(64, (3,3), activation='relu'),
    MaxPooling2D(2,2),
    Flatten(),
    Dense(128, activation='relu'),
    Dense(10, activation='softmax')
])

cnn_cifar.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Train CNN on CIFAR-10
cnn_cifar_history = cnn_cifar.fit(x_train_cifar, y_train_cifar_cat, epochs=10, validation_data=(x_test_cifar, y_test_cifar_cat))

# Plotting function reused
plot_history(ann_cifar_history, "ANN (CIFAR-10)")
plot_history(cnn_cifar_history, "CNN (CIFAR-10)")

# Confusion Matrix and Report - CNN (CIFAR-10)
y_pred_cifar_cnn = np.argmax(cnn_cifar.predict(x_test_cifar), axis=1)
cm_cifar = confusion_matrix(y_test_cifar, y_pred_cifar_cnn)

plt.figure(figsize=(10,8))
sns.heatmap(cm_cifar, annot=True, fmt='d', cmap='Greens')
plt.title("Confusion Matrix - CNN (CIFAR-10)")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()

print("Classification Report - CNN (CIFAR-10):")
print(classification_report(y_test_cifar, y_pred_cifar_cnn))

```
